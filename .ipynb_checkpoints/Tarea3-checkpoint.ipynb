## Parte 1

![](a.png)

Notemos que si los precios de los Fondos se pueden modelar como un MBG, entonces el logaritmo de estos, usando el lema de Ito, se puede modelar como un MBA de la siguiente forma:

$$
\delta f = \left( \mu - \frac{\sigma^2}{2} \right) \cdot \delta t + \sigma \cdot \sqrt{\delta t} \cdot \mathcal{N}(0,1)
$$

Donde $\delta f$ es simplemente $\delta ln(F)$. Podemos reescribir este modelo en versin discreta para pensar en intervalos de das:

$$
\ln(F_{t+\Delta t}) - \ln(F_{t}) = \left( \mu - \frac{\sigma^2}{2} \right) \cdot \Delta t + \sigma \cdot \sqrt{\Delta t} \cdot \mathcal{N}(0,1)
$$

Con esto, la media de los retornos logartmicos diarios sera justamente $\mu_A = \left( \mu - \frac{\sigma^2}{2} \right)$, mientras que la volatilidad sera $\sigma_A = \sigma$, ya que estamos hablando de intervalos de un da; es decir, $\Delta t = 1$.

```python
s = retornos.std()
s
```

```python
mu_sigma2_2 = retornos.mean()
mu = mu_sigma2_2 + s**2/ 2
mu
```

## Parte 2

![](b.png)

Del modelo de MBA presentado en la parte 1 podemos despejar y llegar a la siguiente expresin para el precio de un Fondo:

$$
F_{t+\Delta t} = F_t \cdot \exp \left( \left( \mu - \frac{\sigma^2}{2} \right) \cdot \Delta t + \sigma \cdot \sqrt{\Delta t} \cdot \mathcal{N}(0,1) \right)
$$

Luego, un paso de un mes significa un delta en das de $\Delta t = ''dias\;en\;un\;mes''$, por lo que la simulacin del precio de un Fondo en el ao 35 sera, tomando como precio final el ltimo disponible en los datos y notando que la cantidad de pasos a tomar son 420, ya que es la cantidad de meses 35 aos:

```python
# Precio inicial
F_0 = df.iloc[-1,0]

# Delta meses
n_dias = retornos.shape[0]
Delta_t = n_dias // (5*12) # 5 aos en la base

# Pasos
n_pasos = 12*40

# Distribucin normal estndar simulada
Z = np.random.standard_normal(size=n_pasos)

# Estdsticas del fondo A
s_A = s.iloc[0]
mu_sigma2_2_A = mu_sigma2_2.iloc[0]

# Simulacin del precio recursiva
F_t = F_0 # ltimo precio
F_35 = np.zeros(n_pasos) # precios hata el 35 paso
for t in range(n_pasos):
    F_t = F_t * np.exp(mu_sigma2_2_A * Delta_t + s_A * Delta_t**0.5 * Z[t])
    F_35[t] = F_t # Agregamos el ltimo precio

plt.figure(figsize=(15, 6))
plt.plot(F_35, c='C0')
plt.legend(['FONDO A'])
plt.show()
```

Luego, tenemos que hacer lo mismo para ltimos 5 aos en el fondo E, tomando como precio inicial el ltimo precio de la simulacin anterior, pero con 60 pasos, que es la cantidad de meses en 5 aos y las estadsticas del fondo E.

```python
# Pasos
n_pasos = 60

# Distribucin normal estndar simulada
Z = np.random.standard_normal(size=n_pasos)

# Estdsticas del fondo A
s_B = s.iloc[1]
mu_sigma2_2_B = mu_sigma2_2.iloc[1]

# Simulacin del precio recursiva
F_t = F_35[-1] # ltimo precio
F_5 = np.zeros(n_pasos) # precios hata el 35 paso
for t in range(n_pasos):
    F_t = F_t * np.exp(mu_sigma2_2_B * Delta_t + s_B * Delta_t**0.5 * Z[t])
    F_5[t] = F_t # Agregamos el ltimo precio

plt.figure(figsize=(15, 6))
plt.plot(F_5, c='C1')
plt.legend(['FONDO E'])
plt.show()
```

Ahora combinamos ambos resultados para la simulacin de los 40 aos de cotizaciones en un solo arreglo.

```python
F_40 = np.append(F_35, F_5)

plt.figure(figsize=(15, 6))
plt.plot(F_40, c='C1', label='FONDO E')
plt.plot(F_35, c='C0', label='FONDO A')
plt.legend()
plt.show()
```
